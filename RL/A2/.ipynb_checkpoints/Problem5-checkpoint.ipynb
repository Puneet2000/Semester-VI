```python
import random
states=(['s1','s2','s3'])
actions=(['a1','a2','a3'])
Q=[[0 for i in range(len(actions))] for j in range(len(states))]
rewards=[[i+1 for i in range(len(actions))] for j in range(len(states))]
choice=[0,1,2]
epsilon=0.1
gamma=0.5
alpha=0.7
iterations=10000
epochs=100
for i in range(epochs):
    s=[]
    s_ind=random.choice(choice)
    x=random.random()
    if(x>epsilon):
        a_ind=Q[s_ind].index(max(Q[s_ind]))
    else:
        a_ind=random.choice(choice)
        s.append(states[s_ind])
        s.append(actions[a_ind])
    for j in range(iterations):
        r=rewards[s_ind][a_ind]
        if(r==3):
            s.append('.')
        else:
            s.append(r)
        s_next=a_ind
        a_next=Q[s_next].index(max(Q[s_next]))
        q=Q[s_ind][a_ind]
        Q[s_ind][a_ind]=q+alpha*(r+ gamma*Q[s_next][a_next] -q)
        s_ind=s_next
        x=random.random()
        if(x>epsilon):
            a_ind=Q[s_ind].index(max(Q[s_ind]))
        else:
            a_ind=random.choice(choice)
        if(s_ind==2 ):
            s.append('.')
        elif(a_ind==2):
            s.append('.')
        else:
            s.append(states[s_ind])
            s.append(actions[a_ind])
print(s)
        
print(Q)


# print(states,actions,Q[0].index(max(Q[0])),rewards[:][2])
```
